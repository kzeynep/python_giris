```python
#Python'da bir fonksiyonun parantezleri arasında, fonksiyona geçirilen argümanlar virgülle ayrılarak yazılır. Ancak bazen bir fonksiyona kaç adet argüman geçirileceği önceden bilinmeyebilir. Bu durumda, fonksiyonun tanımında "args" isimli bir değişken kullanarak, fonksiyona geçirilen tüm argümanları toplu bir şekilde işleyebilirsiniz.
def my_function(*args):
    for arg in args:
        print(arg)
```

```python
#"kwargs" değişkeni, bir fonksiyona adlandırılmış argümanlar (keyword arguments) olarak geçirilen, ancak fonksiyon tanımında önceden belirtilmemiş olan argümanları temsil eder. "kwargs" ile belirtilen argümanlar, bir sözlük (dictionary) şeklinde işlenir.
def my_function(**kwargs):
    for key, value in kwargs.items():
        print(key, value)
```

```python
my_function(a=1, b=2, c=3) # a=1, b=2 ve c=3 anahtar kelime argümanları ekrana yazdırılır
my_function(name="John", age=30, country="USA") # name="John", age=30 ve country="USA" anahtar kelime argümanları ekrana yazdırılır
```

```python
#args ve kwargs beraber kullanım da ilk args kullanırız 
def beraber(*args,**kwargs):
  for i in args :
    print(i)

  for i in kwargs.items():
    print(i)

beraber(*range(10))
```

```python
def dış_fonksiyon(*args):
    print("Dış fonksiyon çalışıyor.")

    def iç_fonksiyon_1(*args):
        print("İç fonksiyon 1 çalışıyor.")

        def iç_fonksiyon_2(*args):
            print("İç fonksiyon 2 çalışıyor.")
            print("args:", args)

        iç_fonksiyon_2(*args)

    def iç_fonksiyon_3(*args):
        print("İç fonksiyon 3 çalışıyor.")
        print("args:", args)

    iç_fonksiyon_1(*args)
    iç_fonksiyon_3(*args)

dış_fonksiyon(1, 2, 3)
```

```python
#Decorators: Elimizde var olan fonksiyonları  kolay bir şekilde modifiye etmemizi sağlar ve kod kalabalığını önler.
#Elimizde olan bir fonksiyonu argüman olarak alır ve ve başka bir fonksiyon döndürür.
#Decorator kullanılmadan merhaba kelimesini küçük harflerle yazdık bu fonskiyonu kullanarak merhaba kelimesinin hepsini büyük harfe çevireceğiz

def upper(func):
  def wrapper():
    f = func()
    return f.upper()
  return wrapper 

def selamlama():
  return 'merhaba'
selamlama()
```

```python
def upper(func):
  def wrapper():
    f = func()
    return f.upper()
  return wrapper 
@upper   # bu sayende merhaba kelimemizin bütün harfleri büyük harfe çevrilmiş olacak
def selamlama():
  return 'merhaba'
selamlama()
```

```python
import time 

def mul(*nums):
  start = time.time()
  result = list()
  for num in nums:
    result.append(num*num)

  end = time.time()
  print("Süre : " , str(end-start))
  return result

def add(*nums):
  start = time.time()
  result = list()
  for num in nums:
    result.append(num+num)
  end = time.time()
  print("Süre : " , str(end-start))
  return result 

mul(*range(10000000))
add(*range(10000000))
```

```python
#Decorator fonksiyonu kullanımı 
import time 

def decorator(func):
  def wrapper(*args, **kwargs):
    start = time.time()
    result = func(*args, **kwargs)
    end = time.time()
    print("Süre : ", str(end-start))
    return result
  return wrapper

@decorator
def mul(*nums):
  result = list()
  for num in nums :
    result.append(num*num)
  return result

@decorator
def add(*nums):
  result = list()
  for num in nums :
    result.append(num+num)
  return result 

mul(*range(100000))
add(*range(1000000))
```

```python
#Generator fonksiyonları, bellek kullanımını minimize etmek için kullanılabilir. Normal bir fonksiyon, bir liste veya başka bir veri yapısı döndürmek için tüm öğeleri bellekte tutar. Ancak generator fonksiyonları, veri yapısındaki öğeleri tek tek üretir ve bu nedenle daha az bellek kullanır.
def tek_sayilar():
    i = 1
    while True:
        yield i
        i += 2
```

```python
ts = tek_sayilar()
print(next(ts))  # 1
print(next(ts))  # 3
print(next(ts))  # 5
```

```python
#Generator fonksiyonu kullanmadan yazılan kod :
def func(nums):
  result = list()
  for i in nums:
    result.append(i)
    return result 
x = func(range(10))
print(x)
```

```python
#kullanılmış hali 
def func (nums):
  for i in nums :
    yield i 

gen = func(range(5))
print(gen)
```
