```python
#beautifulsoup html kodu açık olan bir kodunun verilerini çekmeye yardımcı olan bir kod dizimidir.
import requests
from bs4 import BeautifulSoup

url = 'https://www.python.org/'
response = requests.get(url)
#print(response.status_code)
content = response.content
soup = BeautifulSoup(content, 'html.parser') # 'lxml' yerine 'html.parser' kullanıyoruz
print(soup.prettify()) # düzgün biçimlendirilmiş halini yazdırıyoruz


#print(soup.prettify())   #daha düzenli kodlara erişmiş oluruz
print(soup.title)  #title kodunu almak için kullanacağımız kod 
print(soup.title.text) #title içindeki textleri erişmiş olurz.
```
